Check for missing values in the dataset

```python
s= df.isna().sum()
print(s)
```

This is a clean dataset, we can do some feature engineering. We will add which month the data was collected.

```python
df['collected_month'] = df['collected_date'].apply(lambda x: x.split("-")[1])
df['collected_month'] = df['collected_month'].apply(lambda x: "September" if x == "09" else "October")
df['collected_month'].value_counts()
```

Manually extract information from the school name and create new column called school type.

```python
df['school_type'] = df['school'].apply(lambda x: "Public" if ("Public") in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Public" if ("publique") in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Catholic" if ("Catholic") in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Catholic" if ("catholique") in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Elementary" if "Elementary" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Secondary" if "Secondary" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Secondary" if "High" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Middle" if "Middle" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Separate" if "Separate" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Junior" if "Junior" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: "Senior" if "Senior" in x else x)
df['school_type'] = df['school_type'].apply(lambda x: x if len(x.split(" ")) == 1 else "other")



df['school_type'].value_counts()
```

Q1. Which schoolboard has the top 10 highest number of confirmed cases in schools?

```python
top10_schoolboard = df.groupby(['school_board']).sum().sort_values(by = ['total_confirmed_cases'], ascending = False).head(10)
top10_viz=top10_schoolboard.iloc[:,3:4]
top10_viz
```

Q2. Which schoolboards have the top 10 highest number of confirmed student cases?

```python
confirmed_student = df.groupby(['school_board']).sum().sort_values(by = ['confirmed_student_cases'], ascending = False).head(10)
confirmed_student.iloc[:,0:1]
```

Q3. Which schoolboards has the has the top 10 highest number of confirmed staff cases?

```python
confirmed_staff = df.groupby(['school_board']).sum().sort_values(by = ['confirmed_staff_cases'], ascending = False).head(10)
confirmed_staff.iloc[:,1:2]
```

Q4. What type of school has the most confirmed cases?

```python
school_type = df.groupby(['school_type']).sum().sort_values(by = ['total_confirmed_cases'], ascending = False).head(10)
school_type
```

We see that public schools have the highest confirmed student cases. This is probably because there are a lot more students that go to public schools, and also there are a lot more public schools throughout the country compared to private schools. From this information, we know that we need to come up with a protocal for public schools to stop the spread.
